```python
import numpy as np
import seaborn as sns
from sklearn import datasets
from pandas import DataFrame
%pylab inline
```

#### Функция понижения размерности

```python
def down_dimmention(A, n):
    # Понизим размерность через ковариацию
    # Для этого найдём её собственные значения и собственные вектора
    val, vec = np.linalg.eig(np.cov(A.T))
    # Выберем n максимальных собственных значений 
    max_val_ind = np.argsort(val)[::-1][:n]
    # Спроецируем A на собственные векторы для соответствующих максимальных собственных значений 
    A1 = np.dot(A, vec[:, max_val_ind])
    
    print('Нормы Фробениуса:')
    print('До:', np.linalg.norm(A))
    print('После:', np.linalg.norm(A1))
    
    # Сравним визуально
    pp = sns.pairplot(DataFrame(A))
    pp.fig.suptitle("Исходные данные", y=1.1)
    pp = sns.pairplot(DataFrame(A1)) 
    pp.fig.suptitle("Данные после оптимизации", y=1.1)
```

#### Загрузка данных

```python
iris = datasets.load_iris().data
```

#### Уменьшим до размерности 3

```python
down_dimmention(iris, 3)
```

#### Уменьшим до размерности 2

```python
down_dimmention(iris, 2)
```

#### Уменьшим до размерности 1

```python
down_dimmention(iris, 1)
```
